# **YÊU CẦU**

image stitching cho 2 cặp ảnh:
* foto1A.jpg và foto1B.jpg
* match1.jpg và match2.jpg

```python
foto1A="/content/gdrive/MyDrive/CS231.N22.KHCL/Baitap/Baitap8/images/foto1A.jpg"
foto1B="/content/gdrive/MyDrive/CS231.N22.KHCL/Baitap/Baitap8/images/foto1B.jpg"
match_1="/content/gdrive/MyDrive/CS231.N22.KHCL/Baitap/Baitap8/images/match1.jpg"
match_2="/content/gdrive/MyDrive/CS231.N22.KHCL/Baitap/Baitap8/images/match2.jpg"
```

```python
def MATCH(image_path_1, image_path_2):
  image1=cv.imread(image_path_1, 1)
  image2=cv.imread(image_path_2, 1)
  ORB=cv.ORB_create()
  keypoint_1, descriptor_1=ORB.detectAndCompute(image1, None)
  keypoint_2, descriptor_2=ORB.detectAndCompute(image2, None)
  img_1=cv.drawKeypoints(image1, keypoint_1, None, color=(0,255,0), flags=0)
  img_2=cv.drawKeypoints(image2, keypoint_2, None, color=(0,255,0), flags=0)
  bf=cv.BFMatcher(cv.NORM_HAMMING, crossCheck = True)
  matches=bf.match(descriptor_1,descriptor_2)
  matches=sorted(matches, key = lambda x:x.distance)
  final=cv.drawMatches(img_1,keypoint_1,img_2,keypoint_2,matches[:],None)
  plt.figure(figsize=(15,15))
  plt.imshow(cv.cvtColor(final, cv.COLOR_BGR2RGB))
  plt.show()
```

```python
MATCH(foto1A, foto1B)
```

```python
MATCH(match_1, match_2)
```
